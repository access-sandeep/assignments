### Data Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plotPie(columns, size):
    total_size = len(columns)
    ncols = 3
    nrows = int(np.ceil(total_size/3))
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=size)
    
    for i in range(0, nrows):
        for j in range (0, ncols):
            col = columns[i+j]
            e.df[col].value_counts().plot(kind='pie', ax=ax[i][j])
```

```python
plotPie(columns=["BusinessTravel", "Department", "EducationField", "Gender", "StandardHours", "EmployeeCount", "OverTime", "Over18", "MaritalStatus", "JobRole"], size=(15, 20))
```
