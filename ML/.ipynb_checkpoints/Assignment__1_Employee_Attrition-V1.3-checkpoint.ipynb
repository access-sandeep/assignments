### Normal Distribution function:

- x is the point
- mu is mean
- sigma is standard deviation

```python
from math import e, pi, sqrt
```

```python
def nd(x, mu, sig):
    sqrt_towpi = sqrt(2*pi)
    exp = ((x-mu)**2)/(2*sig**2)
    norm = (1/(sig*sqrt_towpi)) * e**(-exp)
    return norm
```

```python
def normal_distribution(X, amplifier=1):
    mu = np.mean(X)
    sig = np.std(X)
    y = [nd(x=x, mu=mu, sig=sig)*amplifier for x in X]
    return y
```

```python
age_series = eadf_numeric["Age"].value_counts().index
age_counts = eadf_numeric["Age"].value_counts().values
```

```python
def impose_normal(column, normal_amplifier=1):
    age_series = eadf_numeric[column].value_counts().index
    age_counts = eadf_numeric[column].value_counts().values
    plt.scatter(age_series, normal_distribution(age_series, normal_amplifier), label="Ideal Normal Distribution", marker='.')
    plt.scatter(age_series, age_counts, label="Actual Distribution", marker='.')
    plt.xlabel(column)
    plt.ylabel(column+" Frequency")
    plt.legend()
    plt.show()    
```

```python
def comparison_plots(index):
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
    ax[0].hist(x=eadf_numeric[hist_numeric_columns[index]])
    ax[0].set_xlabel("Histogram for the "+hist_numeric_columns[index])
    ax[1].boxplot(x=eadf_numeric[hist_numeric_columns[index]])
    ax[1].set_xlabel("Boxplot for the "+hist_numeric_columns[index])
    plt.show()
```

```python
def quartile_report(series):
    name = series.name
    mean = series.mean()
    median = series.median() 
    mode_val = series.mode()
    lower_quartile = series.quantile(0.25) 
    middle_quartile = series.quantile(0.50) 
    upper_quartile = series.quantile(0.75) 
    iqr = upper_quartile-lower_quartile
    lower_whisker = lower_quartile - 1.5 * iqr 
    upper_whisker = upper_quartile + 1.5 * iqr 
    qreport = pd.DataFrame({"Feature":[name], "Mean (Average)":[mean], "Median (Middle)":[median], "Mode (Most repeated)":[mode_val], "Upper Quartile":[upper_quartile], "Middle Quartile":[middle_quartile], "Lower Quartile":[lower_quartile], "IQR":[iqr], "Lower Whisker":[lower_whisker], "Upper Whisker":[upper_whisker]})
    return qreport
```

```python
TN = cm[0,0] #Actually 0, but predicted as 0: TN
FP = cm[0,1] #Actually 0, but predicted as 1 : FP
FN = cm[1,0] #Actually 1, but predicted as 0 : FN
TP = cm[1,1] #Actually 1, but predicted as 1 : TP
N = len(y_test)

accuracy = (TN+TP)/N
recall = TP/(TP+FN)
specificity = TN/(TN+FP)
fpr = 1-specificity
precision = TP/(TP+FP)
f1_score_value = f1_score(y_test, y_pred)
frame_contents = {"Measurement":["Accuracy", "Recall or TPR or Sensitivity", "Specificity", "FPR or 1-Specificity", "Precision", "F1 Score"], "Values": [accuracy, recall, specificity, fpr, precision, f1_score_value]}
df_measurements = pd.DataFrame(frame_contents)
df_measurements
```

```python
fpr, tpr, threshold = roc_curve(y_test,positive_probability)
```

```python
ftt = pd.DataFrame(np.hstack((threshold.reshape(-1,1),fpr.reshape(-1,1),tpr.reshape(-1,1))),columns = ['threshold','fpr','tpr'])
ftt.head(10)
```

```python
ftt.sort_values(by="tpr", ascending=True)
```

```python
def compute_auc(test, train_instance):
    probs = train_instance.predict_proba(test)
    prob_positive = probs[:,1]
    fpr, tpr, threshold = roc_curve(y_test,prob_positive)
    roc_auc = auc(fpr,tpr)
    return roc_auc,fpr,tpr
```

```python
roc_auc, fpr, tpr = compute_auc(X_test,dtc)
roc_auc
```

```python
def plot_ROC(roc,fpr,tpr):
    plt.title("receiver Operating Charecteristc")
    plt.plot(fpr,tpr,'red',label = 'AUC=%0.2f'% roc)
    plt.legend(loc = 'lower right')
    plt.plot([0,1],[0,1], color = 'darkblue',linestyle='--')
    plt.ylabel('True positive Rate')
    plt.xlabel('False positive Rate')
    plt.show()
```

```python
plot_ROC(roc_auc,fpr,tpr)
```
