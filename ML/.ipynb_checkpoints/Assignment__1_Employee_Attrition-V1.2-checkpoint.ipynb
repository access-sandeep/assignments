```python
def draw_hist(columns):
    sclr = StandardScaler()
    eadf_sclaed = sclr.fit_transform(eadf[columns])
    age = eadf_sclaed[:,0]
    attriation = eadf_sclaed[:,1]
    fig, ax = plt.subplots(nrows=13, ncols=2,figsize=(15,100))
    ax[0][0].hist(x=eadf["Age"])
    ax[0][1].hist(x=age)
    ax[0][1].hist(x=attriation)
    ax[1][0].pie(eadf["Attrition"].value_counts(), labels=["Attriation == 0 (No Attriation)", "Attriation == 1 (Attriation)"])
    i = 0
    for row in range(2, 13):
        for col in range(0, 2):
            feature_index = i
            i = i+1
            it = sclr.inverse_transform(eadf_sclaed)
            ax[row][col].hist(x=attriation)
            ax[row][col].hist(x=eadf_sclaed[:,feature_index])
            ax[row][col].set_xlabel("Attriation VRS "+columns[feature_index])
    plt.show()
```

```python
def plotPie(columns, size):
    total_size = len(columns)
    ncols = 3
    nrows = int(np.ceil(total_size/3))
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=size)
    
    for i in range(0, nrows):
        for j in range (0, ncols):
            col = columns[i+j]
            e.df[col].value_counts().plot(kind='pie', ax=ax[i][j])
```

```python
feature_importance = pd.DataFrame({"Features":e.X_train.columns, "Importance":dct_trained.feature_importances_}).sort_values(by="Importance", ascending=False, ignore_index=True)
```

```python
feature_importance["Cumulative Sum"] = feature_importance["Importance"].apply('cumsum')
```

```python
feature_importance["Cumulative Sum"] = (feature_importance["Cumulative Sum"]-np.min(feature_importance["Cumulative Sum"]))/np.std(feature_importance["Cumulative Sum"])
```

```python
feature_importance["Importance"] = (feature_importance["Importance"]-np.min(feature_importance["Importance"]))/np.std(feature_importance["Importance"])
```

```python
feature_importance
```

```python
important_features = []
i = 0;
for feature in feature_importance["Features"]:
    if feature_importance["Importance"][i]>0:
        important_features.append(feature)
    print(feature)
    i = i+1
print(important_features)
```

```python
def compute_auc(test, train_instance):
    probs = train_instance.predict_proba(test)
    prob_positive = probs[:,1]
    fpr, tpr, threshold = roc_curve(y_test,prob_positive)
    roc_auc = auc(fpr,tpr)
    return roc_auc,fpr,tpr
```

```python
roc_auc, fpr, tpr = compute_auc(X_test,dt)
roc_auc
```

```python
def plot_ROC(roc,fpr,tpr):
    plt.title("receiver Operating Charecteristc")
    plt.plot(fpr,tpr,'red',label = 'AUC=%0.2f'% roc)
    plt.legend(loc = 'lower right')
    plt.plot([0,1],[0,1], color = 'darkblue',linestyle='--')
    plt.ylabel('True positive Rate')
    plt.xlabel('False positive Rate')
    plt.show()
```

```python
plot_ROC(roc_auc,fpr,tpr)
```
