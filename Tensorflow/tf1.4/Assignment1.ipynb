```python
import tensorflow as tf
```

```python
tf.__version__
```

```python
sess = tf.Session()
```

```python
graph = tf.get_default_graph()
```

```python
graph.get_operations()
```

```python
A = tf.constant([[256, 879, 231, 495, 367],[959, 333, 757, 300, 456],[234, 907, 806, 867, 375]], name='A')
B = tf.constant([[58, 23,45,34],[45,45,67,98],[56,67,73,60],[87,87,94,46],[59,97,67,71]], name='B')
```

```python
sess.run(A)
```

```python
sess.run(B)
```

```python
C = tf.matmul(A,B, name='C')
```

```python
sess.run(C)
```

```python
D = tf.matmul(B,A, name='D')
```

- **The condition for cross multiplication is Number of columns of first matrix should be equal to number of rows of the second matrix.**
    - A have number of columns = 5
    - B have number of rows = 5
- **However B x A is not possible as the above condition is not true. Meaning,**
    - B have number of columns = 4
    - A have number of rows = 3
- **Therefore matrix multiplication of A x B is possible and above is the result. However, matrix multiplication of B x A is not possible.**
