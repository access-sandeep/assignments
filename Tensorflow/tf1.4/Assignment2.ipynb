```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x=tf.placeholder(tf.float32)
y=tf.placeholder(tf.float32)
```

```python
z = x**3+y**3+20*x**2*y+30*y**2*x+10
```

```python
init=tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)
```

```python
x_input = input("Input value of x ")
y_input = input("Input value of y ")
```

```python
sess.run(z,{x:x_input,y:y_input})
```
